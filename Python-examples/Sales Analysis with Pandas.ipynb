## Reading in the updated dataframe

```python

all_data = pd.read_csv('C:\\Users\\Admin\\Desktop\\datasets\\whole_year_sales.csv')
all_data.head()
```

## Data Cleansing

```python
# to see which rows have NaN value
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

# Drop all the row which have NaN value
all_data = all_data.dropna(how='all') # all means Only those rows which have NaN in all columns will be deleted
all_data.head()
```

#### Find 'Or' and delete it

```python

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float

all_data.head()
```

## What was the best month for sales? how much was the sales?

### Add a column for month

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
# to get the total sale of the particular month
sale = all_data.groupby('Month').sum()['Sales']
print(sale)
```

```python
# ploting the bar chart as per sales
months = range(1,13)

plt.bar(months, sale)
plt.xticks(months)
plt.xlabel('Months (1 - 12)')
plt.ylabel(' Sales in US($)')
plt.show()
```

## Which city had the highest sales?

### Add a city column 

```python
### Example of split
# suppose x = '917 1st St, Dallas, TX 75001'
# output of x.split(',') => ['917 1st St', ' Dallas', ' TX 75001']
# so we want the second item i.e item with 1st index
# perform examples of split to know more

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['city'] = all_data['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')

all_data.head()
```

```python
res = all_data.groupby('city').sum()
res
```

```python
cities = [city for city,df in all_data.groupby('city')]

plt.bar(cities, res['Sales'])
plt.xticks(cities, rotation= 'vertical', size= 8)
plt.xlabel('US Cities')
plt.ylabel('Sales in US($)')
```

### What products are most often sold together

```python
# making a dataframe with all the order ID's who have made more than 1 purchase
df = all_data[ all_data['Order ID'].duplicated(keep=False)]
df.head(10)
```

```python
# listing all the purchase made by a single Order ID
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()

df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
# top items which were bought together as a pair of 2
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common
```

```python
for key,value in count.most_common(15):
    print(key, value)
```

```python
# top items which were bought together as a pair of 3
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
for key,value in count.most_common(15):
    print(key, value)
```
